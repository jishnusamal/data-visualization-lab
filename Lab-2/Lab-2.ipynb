```python
"""
Create a dict of Cricket World Cup winners with year as key & winner as value. Print the name of all best performing countries. 
Display the unique list of countries.
"""

winners = {
    2010: "India",
    2011: "England",
    2012: "India",
    2013: "Pakistan"
}

winners_list = list(winners.values())
winning_freq = {}

for country in winners_list:
    winning_freq[country] = winners_list.count(country)

max_freq = max(winning_freq.values())

print("Best Performing Country")
print([country for country, freq in winning_freq.items() if freq == max_freq][0])
    
print("\n")

print("Unique Countries")
for country in set(winners.values()):
    print(country)
```

```python
# Input a sentence. Count the frequency of each word & store the result in dict. Print the dict.

sen = input("Enter a sentence: ")
sen_list = sen.split()

word_freq = {}

for word in sen_list:
    word_freq[word] = sen_list.count(word)

for (word, freq) in word_freq.items():
    print(f"{word}: {freq} occurences")
```

```python
# Take 2 sets. Perform union, intersection, and difference operations on the sets.
set1 = eval(input("Enter a set of numbers: "))
set2 = eval(input("Enter another set of numbers: "))

print(f"Union: {set1.union(set2)}")
print(f"Intersection: {set1.intersection(set2)}")
print(f"Difference: {set1.difference(set2)}")
```
