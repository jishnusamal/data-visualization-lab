```python
# 1. Find sum, product & division
num1 = int(input("Enter a number 1: "))
num2 = int(input("Enter a number 2: "))

def Sum(a,b):
    return a + b
# Sum = num1 + num2
print(f"Sum is {Sum(num1, num2)}")

diff = num1 - num2
print(f"Difference is {diff}")

prod = num1 * num2
print(f"Product is {prod}")

quot = num1 / num2 if (num2 != 0) else "invalid"
print(f"Quotient is {quot}")

mod = num1 % num2
print(f"Remainder is {mod}")
```

```python
# 2. take an input & call a function to check if the number is prime or not.
num = int(input("Enter a number : "))

def prime(num):
    prime = 0
    for i in range(2, int(sqrt(num))+1):
        # print(i, num%i)
        if num%i == 0:
            prime = 0
            break
        else:
            prime = 1
    return prime

print(f"{num} is {"Prime" if prime(num) == 1 else "Not Prime"}")
```

```python
# 3. Count upto 100 & print all odd numbers & sum of all even numbers

def count_display_sum():
    Sum = 0
    print("The odd numbers are: ")
    for i in range (1, 101):
        if (i%2 == 0):
            Sum += i
        else:
            print(i)
    print(f"Sum of all even numbers is {Sum}.")

count_display_sum()
```

```python
# 5. Count vowels & consonants
def count_vowels_consonants():
    s = input("Enter a string: ")
    vowels = "aeiouAEIOU"
    vowel_n = 0
    consonant_n = 0
    for x in s:
        if x in vowels:
            vowel_n += 1
        else:
            consonant_n += 1
    print(f"Vowels = {vowel_n} \nConsonants = {consonant_n}")

count_vowels_consonants()
```
