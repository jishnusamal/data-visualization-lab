```python
"""
Write a Python function to input two numbers and perform the
Calculator operations of (+, -, *, /). 
"""

num1 = int(input("Enter a number 1: "))
num2 = int(input("Enter a number 2: "))

def Sum(a,b):
    return a + b
# Sum = num1 + num2
print(f"Sum is {Sum(num1, num2)}")

diff = num1 - num2
print(f"Difference is {diff}")

prod = num1 * num2
print(f"Product is {prod}")

quot = num1 / num2 if (num2 != 0) else "invalid"
print(f"Quotient is {quot}")

mod = num1 % num2
print(f"Remainder is {mod}")
```

```python
"""
Write a Python function that takes an integer and returns True if
it's a prime number and False otherwise.
"""

num = int(input("Enter a number : "))

def prime(num):
    prime = 0
    for i in range(2, int(sqrt(num))+1):
        # print(i, num%i)
        if num%i == 0:
            prime = 0
            break
        else:
            prime = 1
    return prime

print(f"{num} is {"Prime" if prime(num) == 1 else "Not Prime"}")
```

```python
"""
Create a Python function that creates a sequence between 1 and
100 and prints all the odd numbers. Compute and display the sum
of all the even numbers.
"""

def count_display_sum():
    Sum = 0
    print("The odd numbers are: ")
    for i in range (1, 101):
        if (i%2 == 0):
            Sum += i
        else:
            print(i)
    print(f"Sum of all even numbers is {Sum}.")

count_display_sum()
```

```python
"""
Write a Python function that takes a string input from the user
and counts the number of vowels and consonants in the string.
"""

def count_vowels_consonants():
    s = input("Enter a string: ")
    vowels = "aeiouAEIOU"
    nv = 0
    nc = 0
    for x in s:
        if not x.isalpha():
            continue
        if x in vowels:
            nv += 1
        else:
            nc += 1
    print(f"Vowels = {nv} \nConsonants = {nc}")

count_vowels_consonants()
```
