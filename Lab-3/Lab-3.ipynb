```python
import numpy as np
```

```python
"""
Generate a 3x4 NumPy array with random integers between 1 & 50.
a. Calculate & print the mean, median and standard deviation
b. Print the sum of all elements & the sum of each row
c. Reshape into a 2x6 array and print it.
"""

arr = np.random.randint(1, 51, size=(3,4))
print(arr)

# Calculate & print the mean, median and standard deviation
print(f"Average: {round(arr.mean(), 2)}\nMedian: {np.median(arr)}\nStandard Deviation: {round(arr.std(), 2)}")

# Print the sum of all elements & the sum of each row
print(f"Sum: {arr.sum()}\nSum of each row: {arr.sum(axis=1)}")

# Reshape into a 2x6 array and print it.
print("\nReshaped", arr.reshape((2,6)), sep="\n")
```

```python
"""
Create 2 3x3 matrices & print it. Perform the following operations:
a. Matrix Addition
b. Matrix Subtraction
c. Matrix Multiplication
d. Transpose
e. Determinant & Inverse
"""

a = np.random.randint(1, 51, size=(3,3))
b = np.random.randint(1, 51, size=(3,3))
print(f"A = {a}\nB = {b}", sep="\n")

# Matrix Addition
print("\nSum", a+b, sep="\n")

# Matrix Subtraction
print("\nDifference", a-b, sep="\n")

# Matrix Multiplication
print("\nMatrix Dot Product", a@b, sep="\n")
print("\nMatrix Elementwise Product", a*b, sep="\n")

# Transpose
print("\nTranspose of A", np.transpose(a), sep="\n")
print("\nTranspose of B", np.transpose(b), sep="\n")

# Determinant
print("\nDeterminant of A", round(np.linalg.det(a), 2), sep="\n")
print("\nDeterminant of B", round(np.linalg.det(b), 2), sep="\n")

# Inverse
print("\nInverse of A", np.linalg.inv(a), sep="\n")
print("\nInverse of B", np.linalg.inv(b), sep="\n")
```
